### 1.2.1 - Tipos de Variáveis

Em Python, existem diversos tipos de variáveis que podem ser usados para armazenar diferentes tipos de dados. Aqui estão alguns dos tipos de variáveis mais comuns:

1. Números inteiros (int):

    São usados para representar números inteiros, como 1, 10, -5, etc. Não possuem casas decimais.
    


2. Números de ponto flutuante (float):

    São usados para representar números com casas decimais, como 3.14, -0.5, etc.

3. Strings (str):

    São usados para representar textos, que podem ser palavras, frases ou até mesmo uma sequência de caracteres.

4. Booleanos (bool):

    São usados para representar valores lógicos Verdadeiro (True) ou Falso (False).

5. Listas (list):

    São usadas para armazenar uma coleção de elementos em uma ordem específica, permitindo adicionar, remover e acessar os elementos.

6. Tuplas (tuple):

    São semelhantes às listas, mas são imutáveis, o que significa que os elementos não podem ser alterados após a criação.

7. Dicionários (dict):

    São usados para armazenar pares de chave-valor, em que cada valor é associado a uma chave única, permitindo acessar os valores usando suas chaves.

### 1.3.2 - Funções Matemáticas

abs(x) - Retorna o valor absoluto de x

pow(x, y) ou x ** y - Retorna x elevado à potência y

round(x, n) - Retorna x arredondado para n dígitos após a vírgula (ou para o inteiro mais próximo se n não for fornecido)

min(iterable) - Retorna o menor valor em iterable

max(iterable) - Retorna o maior valor em iterable

sum(iterable) - Retorna a soma de todos os valores em iterable

```python
# Exemplos
x = -10
y = 5

print(abs(x))  # 10
print(pow(x, 2))  # 100
print(x ** 2)  # 100
print(round(y/3, 2))  # 1.67
print(min([2, 4, 6, 8]))  # 2
print(max([2, 4, 6, 8]))  # 8
print(sum([1, 2, 3, 4, 5]))  # 15
```

### 1.3.3 - Expressões Booleanas e Operadores Lógicos

Uma expressão booleana é uma expressão que pode ser ou verdadeira ou falsa. 

Os exemplos a seguir utilizam o operador ==, que compara dois operandos e produz um resultado positivo True, se forem iguais, e um resultado negativo False, caso contrário.

```python
print(5 == 5)
print(5 == 6)
```

O operador == é um dos operadores de comparação; outros operadores desse tipo
são:

```python
x != y      # x é diferente de y
x > y       # x é maior do que y
x < y       # x é menor do que y
x >= y      # x é maior ou igual a y
x <= y      # x é menor ou igual a y
x is y      # x é o mesmo que y
x is not y  # x não é o mesmo que y
```

Existem três operadores lógicos: and, or, e not. A semântica (significado) desses
operadores são similares ao seu significado em Inglês. Exemplos:



```python
x > 0 and x < 10      # só é verdadeiro se x for maior do que 0 e menor que 10
x%2 == 0 or x%3 == 0  # só é verdadeiro se alguma das condições é verdadeira, isto é se o número é divisível por 2 ou 3
not (x > y)           # só é verdadeiro se x > y é falso; isto é, se x for menor ou igual a y.
```

### 1.3.4 - Execução Condicional  (IF - ELSE - ELIF)

A expressão booleana depois da declaração if é chamada de condição. Nós terminamos essa declaração com símbolo de dois pontos (:) e a(s) linha(s) depois da
declaração if são indentadas.

Se a condição lógica é verdadeira, então a declaração indentada é executada. Se a
condição lógica for falsa, a condição indentada é ignorada.

```python
# Exemplo
if x > 0 :
    print('x é positivo')
```

### Principais métodos para modificar strings

str.upper() - Retorna a string em letras maiúsculas.

str.lower() - Retorna a string em letras minúsculas.

str.capitalize() - Retorna a string com a primeira letra em maiúscula e o resto em minúsculas.

str.title() - Retorna a string com a primeira letra de cada palavra em maiúscula e o resto em minúsculas.

str.strip() - Retorna a string sem espaços em branco à esquerda ou à direita.

str.replace(old, new) - Retorna uma cópia da string com todas as ocorrências de old substituídas por new.

str.startswith(prefix) - Retorna True se a string começa com prefix.

str.endswith(suffix) - Retorna True se a string termina com suffix.

str.split(sep) - Retorna uma lista de substrings separados pelo separador sep.

str.join(iterable) - Concatena todos os elementos em iterable em uma única string separada pelo próprio valor da string original.

```python
# Copiando Listas
'''
Não é possível copiar uma lista simplesmente digitando list2 = list1. 
list2 será apenas uma referência a list1, e alterações feitas em list1 serão feitas automaticamente também em list2.
Existem maneiras de fazer uma cópia, uma delas é usar o método interno List copy().'''

thislist = ["apple", "banana", "cherry"]
mylist = thislist.copy()
print(mylist)

thislist = ["apple", "banana", "cherry"]
mylist = list(thislist)
print(mylist)
```
